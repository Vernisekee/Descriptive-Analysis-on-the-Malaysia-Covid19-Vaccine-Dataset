```python
#Daily number of vaccine administration
plt.figure(figsize=(10,7))
plt.plot(Vax_MY_dataset.date,Vax_MY_dataset.daily_partial, color='m', label='Partial')
plt.plot(Vax_MY_dataset.date,Vax_MY_dataset.daily_full, color='salmon', label='Full')
plt.plot(Vax_MY_dataset.date,Vax_MY_dataset.daily_booster, color='r', label='Booster')
plt.plot(Vax_MY_dataset.date,Vax_MY_dataset.daily, color='c', label='Total')
plt.title("Vaccination (National)", fontsize=25, fontweight=10, color='red')
plt.xlabel("Date", fontsize = 20)
plt.ylabel("Daily Vaccine Administration", fontsize = 20)
plt.legend(loc='upper right')
plt.show()
```


#COVID19 Vaccine Adverse Event Following Immunization (AEFI)

```python
#import dataset 
url = 'https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/vaccination/aefi.csv'
Vax_aefi_dataset = pd.read_csv(url)

#View the first 5 rows
Vax_aefi_dataset.head()
```

```python
#View the last 5 rows  
Vax_aefi_dataset.tail()
```

```python
#View the data types,columns, null value counts etc
Vax_aefi_dataset.info()
```

```python
#View the dimensions of the dataset
print("Vaccination aefi data set dimensions : {}".format(Vax_aefi_dataset.shape))
```

```python
#Data cleaning/transforming
#Remove unrelated columns 
Vax_aefi_dataset.drop(['daily_serious_npra', 'daily_nonserious', 'daily_nonserious_npra', 'daily_nonserious_mysj_dose1', 'daily_nonserious_mysj_dose2'], axis=1, inplace=True)

#Create column to sum aefi for dose 1 and 2 
Vax_aefi_dataset['site_pain'] = Vax_aefi_dataset['d1_site_pain'] + Vax_aefi_dataset['d2_site_pain']
Vax_aefi_dataset['site_swelling'] = Vax_aefi_dataset['d1_site_swelling'] + Vax_aefi_dataset['d2_site_swelling']
Vax_aefi_dataset['site_redness'] = Vax_aefi_dataset['d1_site_redness'] + Vax_aefi_dataset['d2_site_redness']
Vax_aefi_dataset['tiredness'] = Vax_aefi_dataset['d1_tiredness'] + Vax_aefi_dataset['d2_tiredness']
Vax_aefi_dataset['headache'] = Vax_aefi_dataset['d1_headache'] + Vax_aefi_dataset['d2_headache']
Vax_aefi_dataset['muscle_pain'] = Vax_aefi_dataset['d1_muscle_pain'] + Vax_aefi_dataset['d2_muscle_pain']
Vax_aefi_dataset['joint_pain'] = Vax_aefi_dataset['d1_joint_pain'] + Vax_aefi_dataset['d2_joint_pain']
Vax_aefi_dataset['weakness'] = Vax_aefi_dataset['d1_weakness'] + Vax_aefi_dataset['d2_weakness']
Vax_aefi_dataset['fever'] = Vax_aefi_dataset['d1_fever'] + Vax_aefi_dataset['d2_fever']
Vax_aefi_dataset['vomiting'] = Vax_aefi_dataset['d1_vomiting'] + Vax_aefi_dataset['d2_site_pain']
Vax_aefi_dataset['chills'] = Vax_aefi_dataset['d1_chills'] + Vax_aefi_dataset['d2_chills']
Vax_aefi_dataset['rash'] = Vax_aefi_dataset['d1_rash'] + Vax_aefi_dataset['d2_rash']

#Remove columns for aefi dose 1 and 2 since the sum have been calculated
Vax_aefi_dataset.drop(['d1_site_pain', 'd1_site_swelling', 'd1_site_redness', 'd1_tiredness', 'd1_headache', 'd1_muscle_pain',
                       'd1_joint_pain', 'd1_weakness', 'd1_fever', 'd1_vomiting', 'd1_chills', 'd1_rash',
                       'd2_site_pain', 'd2_site_swelling', 'd2_site_redness', 'd2_tiredness', 'd2_headache', 'd2_muscle_pain',
                       'd2_joint_pain', 'd2_weakness', 'd2_fever', 'd2_vomiting', 'd2_chills', 'd2_rash',], 
                      axis=1, inplace=True)
```

```python
#View the cleaned data set
Vax_aefi_dataset
```
